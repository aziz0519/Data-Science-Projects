```python
import pandas as pd

df = pd.read_csv('public-transport-journeys.csv')

journeys = df.average_daily_passenger_journeys

print('public-transport-journeys.csv contains {} rows'.format(len(df)))
print('The maximum number of journeys happened in the year 2014 with {} journeys'.format(max(journeys)))
print('The minimum number of journeys happened in the year 2004 with {} journeys'.format(min(journeys)))
```
